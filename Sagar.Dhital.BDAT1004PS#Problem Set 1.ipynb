using System;

class Program
{
    static void Main(string[] args)
    {
        // Example usage:
        double area = triangleArea(2, 2, 2);
        Console.WriteLine(area);
    }

    // Function to calculate the area of the triangle using Heron's formula
    static double triangleArea(double a, double b, double c)
    {
        // Step 1: Calculate the semi-perimeter
        double s = (a + b + c) / 2;

        // Step 2: Calculate the area using Heron's formula
        double area = Math.Sqrt(s * (s - a) * (s - b) * (s - c));

        // Step 3: Return the calculated area
        return area;
    }
}

using System;

class Program
{
    static void Main(string[] args)
    {
        // Example usage for Part a
        Console.WriteLine(inside(1, 1, 0, 0, 2, 3));  // True
        Console.WriteLine(inside(-1, -1, 0, 0, 2, 3)); // False
    }

    // Function to check if point (x, y) is inside the rectangle defined by (x1, y1) and (x2, y2)
    static bool inside(double x, double y, double x1, double y1, double x2, double y2)
    {
        // Check if the point is within the bounds of the rectangle
        return (x >= x1 && x <= x2 && y >= y1 && y <= y2);
    }
}

```python
def frequencies(text):
    # Define the lowercase alphabet
    letters = 'abcdefghijklmnopqrstuvwxyz'
    
    # Create a list initialized to 0 to count occurrences of each letter
    freq_list = [0] * 26  # List to store frequencies for each letter a-z
    
    # Convert the input text to lowercase to handle case insensitivity
    text = text.lower()
    
    # Loop through each character in the text
    for char in text:
        if char in letters:
            # Find the index of the letter in the alphabet and increment the count
            index = letters.index(char)
            freq_list[index] += 1
    
    return freq_list
```
