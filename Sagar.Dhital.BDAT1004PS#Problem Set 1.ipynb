using System;

class Program
{
    static void Main()
    {
        //Counting the number of letters
        string word = "Supercalifragilisticexpialidocious";
        int letterCount = word.Length;
        Console.WriteLine($"The number of letters in '{word}' is: {letterCount}");

        //Checking if 'Supercalifragilisticexpialidocious' contains 'ice'
        bool containsIce = word.Contains("ice");
        Console.WriteLine($"{containsIce}");

        //Finding the longest word
        string word1 = "Supercalifragilisticexpialidocious";
        string word2 = "Honorificabilitudinitatibus";
        string word3 = "Bababadalgharaghtakamminarronnkonn";

        string longest = word1;
        if (word2.Length > longest.Length)
        {
            longest = word2;
        }
        if (word3.Length > longest.Length)
        {
            longest = word3;
        }

        Console.WriteLine($"{longest}");

        //Sorting composer names alphabetically
        string[] composers = { "Berlioz", "Borodin", "Brian", "Bartok", "Bellini", "Buxtehude", "Bernstein" };

        Array.Sort(composers);

        Console.WriteLine($"The first composer in the dictionary is: {composers[0]}");
        Console.WriteLine($"The last composer in the dictionary is: {composers[composers.Length - 1]}");
    }
}
